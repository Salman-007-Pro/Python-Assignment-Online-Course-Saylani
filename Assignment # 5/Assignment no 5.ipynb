```python
# Question:1
# Write a Python function to calculate the factorial of a number (a non-negative
# integer). The function accepts the number as an argument.
def isInteger(no):
    try:
        no=int(no)
        return no
    except:
        print("Invalide number")
def factorial(no):
    if no==0:
        print("The factorial of Zero(0) is 1")
    if no>0:
        fact=1;
        for i in range(1,no+1):
            fact*=i
        print(f"The factorial of given number is {fact}")
        
    else:
        print("you give the negative number")
no=input("Enter the number: ")
no=isInteger(no)
factorial(no)
```

```python
# Question:2
# Write a Python function that accepts a string and calculate the number of upper
# case letters and lower case letters.
def count_upper_lower(elements):
    ucount=0
    lcount=0
    for i in elements:
        if i.isupper():
            ucount+=1
        elif i.islower():
            lcount+=1
        else:
            pass
    print(f"The Uppercase letter: {ucount}\nThe Lowercase letter: {lcount}")
elements=input("enter the string: ")
count_upper_lower(elements)
```

```python
# Question:3
# Write a Python function to print the even numbers from a given list.
lst=[i for i in range(0,10)]
def even(lst):
    for i in lst:
        if i%2==0:
            print(f"The even number is {i}")
even(lst)
```

```python
# Question:4
# Write a Python function that checks whether a passed string is palindrome or not.
# Note: A palindrome is a word, phrase, or sequence that reads the same
# backward as forward, e.g., madam
element=input("enter the word: ")
element=element.lower()
def palindrome(word):
    word=word[::-1]
    if word==element:
        print("Its is a Palindrome")
    else:
        print("Its is not a Palindrome")
palindrome(element)
```

```python
# Question:5
# Write a Python function that takes a number as a parameter and check the
# number is prime or not.
def isInteger(no):
    try:
        no=int(no)
        return no
    except:
        print("Invalide number")
def isnegative(no):
    if no<=0:
        return no*-1
    else:
        return no
def isprime(no):
    flag=False
    for i in range(2,no):
        if no%i==0:
            flag=True
            break
    if not flag:
        print(f"The given number is prime({no}) number")
    else:
        print(f"The give number is not prime({no}) number")
no=int(input("enter the number: "))
no=isInteger(no)
no=isnegative(no)
# print(type(no))
isprime(no)
```
